```python
def createGenerator():
    mylist = range(3)
    for i in mylist:
        yield i * i
        
mygenerator = createGenerator()  # 제너레이터 생성
print(mygenerator)  # mygenerator는 객체입니다.
```

```python
for i in mygenerator:
    print(i)
```

```python
class Bank(): # 은행을 맏을고 ATM도 만듭시다.
    crisis = False
    def create_atm(self):
        while not self.crisis:
            yield "$100"
```

```python
hsbc = Bank()  # 아무 문제 없다면 ATM은 원하는 만큼 줄겁니다.
```

```python
corner_street_atm = hsbc.create_atm()
```

```python
print(corner_street_atm.__next__())  # in python 2.x version in .next() instead of .__next__()
```

```python
print(corner_street_atm.__next__())
```

```python
print([corner_street_atm.__next__() for cash in range(5)])
```

```python
hsbc.crisis = True  # 경제 공항이 오고 있습니다, 돈이 더 없겠네요
```

```python
print(corner_street_atm.__next__())
```

```python
wall_street_atm = hsbc.create_atm()  # 새로운 ATM을 만들어도 돈은 여전히 없겠죠
```

```python
print(wall_street_atm.__next___())
```

```python
hsbc.crisis = False  # 경제 공황이 끝났음에도 ATM은 모두 비어있습니다.
```

```python
print(corner_street_atm.__next__())
```

```python
brand_new_atm = hsbc.create_atm()  # 다시 일상으로 돌아가기 위해
```

```python
for cash in brand_new_atm:
    print(cash)
```
